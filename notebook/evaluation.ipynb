```python
import pandas as pd
```

```python
#input files
ranked_df = pd.read_csv("../data/ranked_resumes.csv")
skill_gap_df = pd.read_csv("../data/skill_gap_analysis.csv")

ranked_df.head()
```

```python
#inspect top jobs
job_id = ranked_df["job_id"].iloc[0]
top_resumes = ranked_df[ranked_df["job_id"] == job_id]

top_resumes.head()
```

```python
#Combine ranking and skill gap
evaluation_df = pd.merge(
    ranked_df,
    skill_gap_df,
    on=["job_id", "resume_id"],
    how="left"
)

evaluation_df.head(100)
```

```python
#evaluate quality
top_k = 3

top_k_df = evaluation_df[evaluation_df["rank"] <= top_k]
top_k_df.head()
```

```python
#sanity check
row = top_k_df.iloc[0]

print("Job ID:", row["job_id"])
print("Resume ID:", row["resume_id"])
print("Skill Match %:", row["skill_match_percent"])
print("Missing Skills:", row["missing_skills"])
```

```python
evaluation_df.to_csv("../data/evaluation_results.csv", index=False)
print("Evaluation results saved")
```
