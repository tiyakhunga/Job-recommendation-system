```python
import pandas as pd
```

```python
#Load JD Excel
def load_jobs_from_excel(file_path):
    df = pd.read_excel(file_path)

    jobs = {}
    for _, row in df.iterrows():
        job_text = f"""
        Job Title: {row['Job Title']}
        Job Description: {row['Job Description']}
        """
        jobs[f"job_{row['S.No']}"] = job_text.strip()

    return jobs
```

```python
#print JD
jobs = load_jobs_from_excel("../data/JD.xlsx")

for name, text in jobs.items():
    print(f"\n {name} \n")
    print(text[:500])
```
