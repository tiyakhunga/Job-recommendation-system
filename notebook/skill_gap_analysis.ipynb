```python
import pandas as pd
import ast
```

```python
#input files
resumes_df = pd.read_csv("../data/parsed_resumes.csv")
jobs_df = pd.read_csv("../data/parsed_jobs.csv")

resumes_df.head(1)
```

```python
#Convert skill strings back to lists#
resumes_df["skills"] = resumes_df["skills"].apply(ast.literal_eval)
jobs_df["required_skills"] = jobs_df["required_skills"].apply(ast.literal_eval)
```

```python
#Skill gap function
def skill_gap_analysis(resume_skills, job_skills):
    resume_set = set(resume_skills)
    job_set = set(job_skills)

    matched = resume_set.intersection(job_set)
    missing = job_set - resume_set

    if len(job_set) == 0:
        match_percent = 0
    else:
        match_percent = round(len(matched) / len(job_set) * 100, 2)

    return match_percent, list(missing)
```

```python
#Compare every resume with every JD
results = []

for _, job in jobs_df.iterrows():
    for _, resume in resumes_df.iterrows():
        match_percent, missing_skills = skill_gap_analysis(
            resume["skills"], job["required_skills"]
        )

        results.append({
            "job_id": job["job_id"],
            "resume_id": resume["resume_id"],
            "skill_match_percent": match_percent,
            "missing_skills": missing_skills
        })
```

```python
#Skill gap dataframe 
skill_gap_df = pd.DataFrame(results)
skill_gap_df.head(100)
```

```python
skill_gap_df.to_csv("../data/skill_gap_analysis.csv", index=False)
print("Skill gap analysis saved")
```
