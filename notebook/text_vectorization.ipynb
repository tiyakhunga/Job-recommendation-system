```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
resumes_df = pd.read_csv("../data/resumes_cleaned.csv")
jobs_df = pd.read_csv("../data/jobs_cleaned.csv")

# Quick check
print(resumes_df.head(1))
print(jobs_df.head(1))
```

```python
# Initialize vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit on all resume + JD text together (optional: ensures same feature space)
all_text = pd.concat([resumes_df["clean_text"], jobs_df["clean_text"]])

tfidf_vectorizer.fit(all_text)

# Transform resumes
resume_tfidf = tfidf_vectorizer.transform(resumes_df["clean_text"])

# Transform JDs
jd_tfidf = tfidf_vectorizer.transform(jobs_df["clean_text"])

print("Resume TF-IDF shape:", resume_tfidf.shape)
print("JD TF-IDF shape:", jd_tfidf.shape)
```

```python
# Feature names (words)
features = tfidf_vectorizer.get_feature_names_out()
print("First 20 features:", features[:20])

# Sample TF-IDF vector for first resume
print(resume_tfidf[0].toarray()[0][:20])
```
