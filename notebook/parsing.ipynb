```python
import pandas as pd
import re
```

```python
resumes_df = pd.read_csv("../data/resumes_cleaned.csv")
jobs_df = pd.read_csv("../data/jobs_cleaned.csv")

resumes_df.head(1)
```

```python
#Create a Skill Dictionary
PROGRAMMING_LANGUAGES = [
    "python", "java", "c", "c++", "c#", "javascript",
    "typescript", "go", "rust", "r", "scala", "kotlin",
    "php", "swift", "dart", "bash"
]
DATA_SKILLS = [
    "sql", "mysql", "postgresql", "mongodb",
    "data analysis", "data analytics", "data visualization",
    "excel", "power bi", "tableau", "statistics",
    "pandas", "numpy", "matplotlib", "seaborn"
]
ML_AI_SKILLS = [
    "machine learning", "deep learning", "artificial intelligence",
    "supervised learning", "unsupervised learning",
    "nlp", "computer vision", "time series",
    "feature engineering", "model training", "model evaluation",
    "scikit-learn", "tensorflow", "pytorch", "keras"
]
GENAI_SKILLS = [
    "llm", "large language model", "generative ai",
    "prompt engineering", "openai api",
    "langchain", "llamaindex", "rag",
    "vector database", "embeddings", "faiss", "pinecone"
]
BACKEND_SKILLS = [
    "backend", "rest api", "api development",
    "django", "flask", "fastapi",
    "spring boot", "nodejs", "express",
    "microservices", "authentication", "authorization"
]
FRONTEND_SKILLS = [
    "html", "css", "javascript", "react",
    "angular", "vue", "nextjs",
    "bootstrap", "tailwind", "ui", "ux"
]
DATABASE_BIGDATA_SKILLS = [
    "database", "sql server", "oracle",
    "redis", "elasticsearch",
    "hadoop", "spark", "pyspark",
    "kafka", "airflow", "etl"
]
CLOUD_DEVOPS_SKILLS = [
    "aws", "azure", "gcp",
    "docker", "kubernetes",
    "ci/cd", "jenkins",
    "terraform", "ansible",
    "linux", "unix"
]
SOFTWARE_PRACTICES = [
    "git", "github", "gitlab",
    "version control",
    "unit testing", "integration testing",
    "agile", "scrum", "jira",
    "design patterns", "system design"
]
DATA_ENGINEERING_SKILLS = [
    "data pipeline", "data warehouse",
    "bigquery", "redshift", "snowflake",
    "airflow", "dbt", "etl", "elt"
]
SECURITY_MISC_SKILLS = [
    "cybersecurity", "encryption",
    "oauth", "jwt",
    "performance optimization",
    "scalability", "monitoring"
]
```

```python
SKILLS = (
    PROGRAMMING_LANGUAGES
    + DATA_SKILLS
    + ML_AI_SKILLS
    + GENAI_SKILLS
    + BACKEND_SKILLS
    + FRONTEND_SKILLS
    + DATABASE_BIGDATA_SKILLS
    + CLOUD_DEVOPS_SKILLS
    + SOFTWARE_PRACTICES
    + DATA_ENGINEERING_SKILLS
    + SECURITY_MISC_SKILLS
)
```

```python
#Extracting skills found resume
def extract_skills(text, skill_list):
    found_skills = []
    for skill in skill_list:
        if skill in text:
            found_skills.append(skill)
    return found_skills
```

```python
#Create a education Dictionary
EDUCATION_KEYWORDS = [

    # Undergraduate Degrees
    "btech", "b.tech", "be" ,"b.e",
    "bsc", "b.sc", "bca",
    "bachelor of technology",
    "bachelor of engineering",
    "bachelor of science",
    "bachelor of computer applications",

    # Postgraduate Degrees
    "mtech", "m.tech", "me" ,"m.e",
    "msc", "m.sc", "mca",
    "master of technology",
    "master of engineering",
    "master of science",
    "master of computer applications",

    # Doctorate
    "phd", "ph.d", "doctorate",

    # Diplomas
    "diploma",
    "polytechnic",

    # Computer Science / Tech Specializations
    "computer science",
    "information technology",
    "information science",
    "software engineering",
    "data science",
    "artificial intelligence",
    "machine learning",
    "cyber security",
    "cloud computing",
    "big data",
    "computer engineering",
    "electronics and communication",
    "electronics engineering",
    "electrical engineering",

    # Certifications / Academic Signals
    "certification",
    "certified",
    "coursework",
    "specialization",
    "nanodegree",

    # Institutions (generic)
    "university",
    "college",
    "institute",
    "school of engineering",
    "school of computer science"
]
```

```python
#extract education found in resume
def extract_education(text):
    found_education = []
    for edu in EDUCATION_KEYWORDS:
        if edu in text:
            found_education.append(edu)
    return list(set(found_education))
```

```python
#extract experince found in resume
def extract_experience(text):
    match = re.search(r'(\d+)\s+year', text)
    if match:
        return match.group(1)
    return "Not mentioned"
```

```python
#extract required skills from JD
jobs_df["required_skills"] = jobs_df["clean_text"].apply(
    lambda x: extract_skills(x, SKILLS)
)
```

```python
#extract education requirement from JD
jobs_df["education_required"] = jobs_df["clean_text"].apply(extract_education)
```

```python
#APPLY parsing on resume
resumes_df["skills"] = resumes_df["clean_text"].apply(
    lambda x: extract_skills(x, SKILLS)
)

resumes_df["education"] = resumes_df["clean_text"].apply(extract_education)
resumes_df["experience"] = resumes_df["clean_text"].apply(extract_experience)
```

```python
#APPLY parsing on JD
jobs_df["required_skills"] = jobs_df["clean_text"].apply(
    lambda x: extract_skills(x, SKILLS)
)

jobs_df["education_required"] = jobs_df["clean_text"].apply(extract_education)

```
