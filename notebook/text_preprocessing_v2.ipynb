```python
import re
import pandas as pd
import nltk

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
resumes_df = pd.read_csv("../data/resumes_raw.csv")
jobs_df = pd.read_csv("../data/jobs_raw.csv")
```

```python
print(resumes_df.loc[0, "raw_text"][:400])
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def clean_text(text):
    """
    Cleans and normalizes resume / job description text.
    """

    # 1. Convert to lowercase
    text = text.lower()

    # 2. Remove email addresses
    text = re.sub(r'\S+@\S+', ' ', text)

    # 3. Remove URLs
    text = re.sub(r'http\S+|www\S+', ' ', text)

    # 4. Remove numbers
    text = re.sub(r'\d+', ' ', text)

    # 5. Remove punctuation and special characters
    text = re.sub(r'[^a-z\s]', ' ', text)

    # 6. Tokenize (split into words)
    tokens = text.split()

    # 7. Remove stopwords and lemmatize
    cleaned_tokens = [
        lemmatizer.lemmatize(word)
        for word in tokens
        if word not in stop_words and len(word) > 2
    ]

    # 8. Join tokens back to string
    return " ".join(cleaned_tokens)
```

```python
resumes_df["clean_text"] = resumes_df["raw_text"].apply(clean_text)
jobs_df["clean_text"] = jobs_df["raw_text"].apply(clean_text)
```
